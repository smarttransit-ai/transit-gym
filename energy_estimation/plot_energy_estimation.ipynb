```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
energy_folder=' '
```

```python
trip=pd.read_csv(f'{energy_folder}/diesel_Trajectory_Route1_ALTON_PARK_block105_trip151688020.csv.csv') 
```

```python
def conv_time(ms):
    s = ms/1000
    time = str(datetime.timedelta(seconds=s))
    return time
```

```python
trip['time']=trip['time_ms'].map(conv_time)
```

```python
#plot
fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(trip.index, trip.energy_gal, color='blue', linewidth=1)
ax1.set_xticks(trip.index[::360])
ax1.set_xticklabels(trip.time[::360], rotation=45)
plt.tight_layout(pad=4)
plt.subplots_adjust(bottom=0.15)
plt.xlabel("Time",fontsize = 14)
plt.ylabel("Estimated energy consumption (gal)",fontsize = 14)
plt.show()
```
